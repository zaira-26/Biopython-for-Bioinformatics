```python
from Bio.Align import PairwiseAligner
```

```python
# Match, Mismatch and Gaps

# Sequence alignment compares two sequences character by character.
# During alignment, three situations can occur:
# - Match: identical nucleotides or amino acids
# - Mismatch: different residues
# - Gap: insertion or deletion introduced to maximize similarity

# The alignment score depends on how these events are rewarded or penalized.
```

```python
aligner = PairwiseAligner()

seq1 = "ATCGTAGCCTAGCGA"
seq2 = "ATCGTACCCTAGCGA"

alignments = aligner.align(seq1, seq2)

print(alignments[0])
print("Score:", alignments.score)
```

```python
aligner = PairwiseAligner()

seq1 = "AAGCTAAA"
seq2 = "AAGCTCCC"

alignments = aligner.align(seq1, seq2)

print(alignments[0])
print("Score:", alignments.score)
```

```python
aligner = PairwiseAligner()

seq1 = "ATCGTAGCCTAGCGA"
seq2 = "ATCGTACCTAGCGA"

alignments = aligner.align(seq1, seq2)

print(alignments[0])
print(alignments.score)
print(len(alignments))
print(type(alignments[0]))
print(dir(alignments[0]))
```

```python
# There are two types of alignments:

# 1. Local alignment: it identifies the most similar region shared between two sequences rather than aligning them from end to end.
# This method is useful when only conserved regions are expected to match.

# 2. Global alignment: it attempts to align the entire length of both sequences.
# This approach is most suitable when the sequences are of similar length and are expected to share high overall similarity.
```

```python
aligner = PairwiseAligner()

aligner.mode = "local"

seq1 = "AAAAATCGTAGCC"
seq2 = "ATCGTAG"
alignments = aligner.align(seq1, seq2)

print(alignments[0])
print(alignments.score)
```

```python
aligner = PairwiseAligner()

aligner.mode = "global"

seq1 = "AAAAATCGTAGCC"
seq2 = "ATCGTAG"

alignments = aligner.align(seq1, seq2)

print(alignments[0])
print(alignments.score)
```

```python
# Alignment scoring

# The quality of an alignment depends on the scoring scheme.
# Common scoring parameters include:
# - Match score
# - Mismatch penalty
# - Gap opening penalty
# - Gap extension penalty

# Adjusting these values changes the resulting alignment.
```

```python
from Bio.Align import PairwiseAligner

aligner = PairwiseAligner()

aligner.match_score = 2
aligner.mismatch_score = -1
aligner.open_gap_score = -3
aligner.extend_gap_score = -0.5

seq1 = "ATCGTAG"
seq2 = "ATCGCAG"

alignments = aligner.align(seq1, seq2)

print(alignments[0])
print("Score:", alignments.score)
```

```python
# Identity calculation using zip function
```

```python
aligner = PairwiseAligner()

seq1 = "ATGTAATAGA"
seq2 = "ATCATTTAGA"

alignments = aligner.align(seq1, seq2)
print(alignments[0])

matches = 0

for a, b in zip(seq1, seq2):
  if a == b:
    matches += 1

identity = (matches / len(seq1)) * 100
print("Identity percentage:", identity, "%")
```

```python
aligner = PairwiseAligner()

seq1 = "ATGCGATCGGAATTGCGGATGCTA"
seq2 = "ATGCGATCCGAATTGCGGATGCTA"

alignments = aligner.align(seq1, seq2)
print(alignments[0])

matches = 0

for a, b in zip(seq1, seq2):
  if a == b:
    matches += 1

identity = (matches/len(seq1)) * 100

print(identity)
```

```python
# Practice questions
```

```python
# A researcher has sequenced a patient's gene and wants to compare it with the reference sequence.
# Perform a global alignment.
# Print the best alignment.
# Print the alignment score.
# Calculate the percentage identity.

aligner = PairwiseAligner()
aligner.mode = "global"

ref_seq = "ATGCGTACGGTA"
patient_seq = "ATGCGATGGGTA"

alignments = aligner.align(ref_seq, patient_seq)

print(alignments[0])
print("Score:", alignments.score)

matches = 0
for a, b in zip(ref_seq, patient_seq):
  if a == b:
    matches += 1
identity = (matches/len(ref_seq)) * 100
print("Identity percentage:", identity)
```

```python
# A small DNA fragment may be part of a larger gene
# Perform a global alignment.
# Perform a local alignment.
# Print both alignments.
# Print both scores.

aligner = PairwiseAligner()
larger_seq = "AAAAATCGTAGCCGGGGG"
fragment = "ATCGTAG"

aligner.mode = "global"
alignments = aligner.align(larger_seq, fragment)
print(alignments[0])
print("Score:", alignments.score)

aligner.mode = "local"
alignments = aligner.align(larger_seq, fragment)
print(alignments[0])
print("Score:", alignments.score)

# Local alignment is best for this question
```

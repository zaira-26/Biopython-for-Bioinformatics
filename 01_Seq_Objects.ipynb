```python
# Importing the library
```

```python
from Bio.Seq import Seq
```

```python
# Finding length of sequence
# Indexing
# Slicing
# Counting bases
```

```python
dna = Seq("ATCGACGTAA")
print(dna)
print(len(dna))
print(dna[1])
print(dna[0:5])
print(dna.count("A"))
print(dna.count("T"))
print(dna.count("C"))
print(dna.transcribe())
print(dna.reverse_complement())
```

```python
# Concatenation
```

```python
seq1 = Seq("ATGCGATTTGCGG")
seq2 = Seq("ATGCGGTCA")
print(seq1 + seq2)
print("GCAT" in seq1)           # membership test
```

```python
# Back transcribe: converting RNA to DNA
```

```python
rna = Seq("AUGACGGUUCAGAUC")
print(rna)
print("DNA sequence:", rna.back_transcribe())
```

```python
# Translation
```

```python
protein = Seq("MKTLLV")
print(protein)
```

```python
# Practice questions
```

```python
dna_seq = Seq("ATGGCTAACCGTTAGC")
print(dna_seq)
print("The total length of DNA sequence is:", len(dna_seq))
print("The first nucleotide is:", dna_seq[0])
print("The last nucleotide is:", dna_seq[-1])
print(dna_seq[4:11])
print("Number of A in the sequence:", dna_seq.count("A"))
print("Number of G in the sequence:", dna_seq.count("G"))
print("AAC" in dna_seq)

dna_seq2 = Seq("GGGCCC")
print("Combined DNA sequence:", dna_seq + dna_seq2)

percent_a = dna_seq.count("A") / len(dna_seq) * 100
print("Percentage of A:", percent_a, "%")
```

```python
dna = Seq("ATGGCTAACCGTTAG")
print("Orginal DNA sequence:", dna)
print("DNA complement:", dna.complement())
print("Reverse complement:", dna.reverse_complement())
print("RNA sequence:", dna.transcribe())
print("Protein:", dna.translate())
```

```python
# Translation with stop codon
```

```python
dna = Seq("ATGTAAATGCCC")
print("Protein:", dna.translate())
print("Protein:", dna.translate(to_stop = True))
```

```python
dna = Seq("ATGTTTAAA")
print(dna.translate())
```

```python
# Sequence statistics
```

```python
from Bio import SeqIO

count = 0

for record in SeqIO.parse("practice.fasta", "fasta"):
  print("Gene:", record.id)
  print("Sequence:", record.seq)
  print("Length:", len(record.seq))
  print("A:", record.seq.count("A"))
  print("G:", record.seq.count("G"))
  print("C:", record.seq.count("C"))
  print("T:", record.seq.count("T"))
  print("GC percentage:", (record.seq.count("G") + record.seq.count("C"))/ len(record.seq) * 100, "%")
  print("AT percentage:", (record.seq.count("A") + record.seq.count("T")) / len(record.seq) * 100, "%")
  print()

  count += 1

print("Total number of genes:", count)
```

```python
# GC and AT content analysis
```

```python
highest_gc = 0
highest_gene = ""

for record in SeqIO.parse("practice.fasta", "fasta"):
  seq = record.seq

  G = seq.count("G")
  C = seq.count("C")

  gc = (G+C)/ len(seq) * 100
  if gc > highest_gc:
    highest_gc = gc
    highest_gene = record.id

print(highest_gene)
print(highest_gc)
```

```python
# Splitting FASTA files

for record in SeqIO.parse("practice.fasta", "fasta"):
  filename = record.id + ".fasta"

  SeqIO.write(record, filename, "fasta")
  print("Created:", filename)
```

```python
# Removing duplicate sequences
# we create set for this as sets cannot contain duplicates

detected_sequences = set()
unique_records = []

for record in SeqIO.parse("practice.fasta", "fasta"):
  if record.seq not in detected_sequences:
    detected_sequences.add(record.seq)
    unique_records.append(record)

SeqIO.write(unique_records, "unique_sequences.fasta", "fasta")
print("Number of unique sequences:", len(unique_records))
```

```python
# Sorting FASTA records
```

```python
# Sorting sequences by length (smallest to largest)

records = list(SeqIO.parse("practice.fasta", "fasta"))
sorted_records = sorted(records, key =lambda record: len(record.seq))

SeqIO.write(sorted_records, "sorted_sequences.fasta", "fasta")
for record in sorted_records:
  print(record.id)
```

```python
# Sorting sequences by length (largest to smallest)

records = list(SeqIO.parse("practice.fasta", "fasta"))
sorted_records = sorted(records, key =lambda record: len(record.seq), reverse = True)

SeqIO.write(sorted_records, "sorted_sequences.fasta", "fasta")
for record in sorted_records:
  print(record.id)
```

```python
# Sorting the sequences alphabetically

records = list(SeqIO.parse("practice.fasta", "fasta"))
sorted_records = sorted(records, key = lambda record: record.id)

SeqIO.write(sorted_records, "sorted_sequences.fasta", "fasta")

for record in sorted_records:
  print(record.id)
```

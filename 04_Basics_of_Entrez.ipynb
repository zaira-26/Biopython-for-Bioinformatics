```python
# Retrieving Records
# After obtaining record IDs from a search, the `efetch()` function is used to download complete records in various formats such as GenBank, FASTA or XML.

# The `esearch()` function searches an NCBI database using a query term and returns the identifiers (IDs) of matching records.
```

```python
genbank_id = "NC_001477"

with Entrez.efetch(db="nucleotide", id=genbank_id, rettype="gb", retmode="text") as handle:
    record = SeqIO.read(handle, "genbank")
```

```python
# Exploring record information
# Each GenBank record contains valuable metadata including:
# - Sequence ID
# - Description
# - Organism
# - Sequence length
# - Biological annotations
```

```python
print("Sequence ID:", record.id)
```

```python
print("Sequence Description:", record.description)
```

```python
print("Sequence Length:", len(record.seq))
```

```python
print("Sequence (first 50 bases):", record.seq[:50])
```

```python
print(record.format("fasta"))
```

```python
# Accessing the sequence

sequence = record.seq
g_content = sequence.count("G")
c_content = sequence.count("C")
gc_content = g_content + c_content
gc_percentage = (gc_content / len(sequence)) * 100

print(f"G count: {g_content}")
print(f"C count: {c_content}")
print(f"GC content: {gc_content}")
print(f"GC percentage: {gc_percentage:.2f}%")
```

```python
#To calculate the percentage of adenine (A) and thymine (T) bases in the sequence

a_content = sequence.count("A")
t_content = sequence.count("T")
at_content = a_content + t_content
at_percentage = (at_content / len(sequence)) * 100
print(f"AT percentage: {at_percentage:.2f}%")
print(f"A count: {a_content}")
print(f"T count: {t_content}")
print(f"AT content: {at_content}")
```

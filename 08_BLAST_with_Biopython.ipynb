```python
# What is BLAST?
# BLAST (Basic Local Alignment Search Tool) is one of the most widely used bioinformatics tools for identifying similar biological sequences. It compares a query sequence against a database and reports statistically significant matches.

# BLAST is commonly used for:
# - Gene identification
# - Species identification
# - Homology searches
# - Functional annotation
```

```python
# Understanding BLAST result:
# 1. Identity : this explains the percentage of aligned bases which are similar; high identity = similar sequence

# 2. Query coverage : this answers how much the sequence was aligned.

# 3. E-value : this answers the likeliness of the match to happen randomly; 0 = extremely strong; 1e-100 = very strong; 1e-10 = good, 0.1 = weak; 10 = poor
```

```python
from Bio.Blast import NCBIWWW
```

```python
# Running a BLAST Search

# Biopython provides the `NCBIWWW.qblast()` function to submit sequences directly to the NCBI BLAST server.
# The function returns an XML file containing detailed alignment information.
```

```python
sequence = "ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG"     # pratice sequence
result_handle = NCBIWWW.qblast("blastn", "nt", sequence)

with open("blast_result.xml", "w") as file:
  file.write(result_handle.read())

print("BLAST completed")
```

```python
# Reading BLAST Results

# BLAST results are commonly stored in XML format.
# Biopython's `NCBIXML` parser converts these XML files into Python objects that can be explored programmatically.
```

```python
from Bio.Blast import NCBIXML

result_file = open("blast_result.xml")
blast_record = NCBIXML.read(result_file)
print(blast_record.application)
print(blast_record.query)
print(blast_record.alignments)
print(len(blast_record.alignments))
```

```python
import os
print(os.listdir())
print(os.path.getsize("TP53_blast.xml"))
```

```python
from Bio.Blast import NCBIXML     # Read XML

with open("TP53_blast.xml") as result_handle:
  blast_record = NCBIXML.read(result_handle)

print(blast_record.application)
print(blast_record.query)
print(blast_record.query_length)
print(blast_record.alignments)    # Each alignment is one BLAST hit
print(len(blast_record.alignments))
print(blast_record.alignments[0])
```

```python
top_hit = blast_record.alignments[0]
print(top_hit.title)
print(top_hit.length)
print(top_hit.hsps)
```

```python
# High-scoring Segment Pair (HSP):

# An HSP represents the highest-scoring aligned region between the query sequence and a database sequence.
# Each alignment may contain one or more HSPs.
```

```python
hsp = top_hit.hsps[0]
print("Score:", hsp.score)
print("E-value:", hsp.expect)
print("Alignment length:", hsp.align_length)
print("Identities:", hsp.identities)
```

```python
# Sequence Identity

# Sequence identity measures the percentage of identical residues between the aligned sequences.
# Higher identity generally indicates greater evolutionary similarity.
```

```python
identity = (hsp.identities/hsp.align_length) * 100
print("Identity percentage:", identity)
```

```python
# Looping through BLAST hits
```

```python
for alignment in blast_record.alignments[:5]:
  print("Title:", alignment.title)
  print("Length:", alignment.length)
  print("-" * 60)
```

```python
# Extracting hits, identity, coverage and E-value for top 5 aligned sequences
```

```python
for alignment in blast_record.alignments[:5]:

  hsp = alignment.hsps[0]

  print("Title:", alignment.title)
  print("Subject length:", alignment.length)
  print("Alignment length:", hsp.align_length)
  print("Score:", hsp.score)
  print("E-value:", hsp.expect)
  print("Identity:", round((hsp.identities/hsp.align_length) * 100,2), "%")

```

```python
# Exporting BLAST Results

# Saving BLAST output as a CSV file makes it easier to compare hits, create reports and perform downstream analyses in spreadsheet software or Python.
```

```python
# Creating CSV report for top 10 alignments
```

```python
import csv

with open("blast_report.csv", "w", newline="") as file:

    writer = csv.writer(file)

    writer.writerow([
        "Hit",
        "Title",
        "Identity %",
        "E-value",
        "Score",
        "Alignment Length"
    ])

    for i, alignment in enumerate(blast_record.alignments[:10], start=1):

        hsp = alignment.hsps[0]
        identity = (hsp.identities / hsp.align_length) * 100

        writer.writerow([
            i,
            alignment.title,
            round(identity, 2),
            hsp.expect,
            hsp.score,
            hsp.align_length
        ])

print("CSV report saved!")
```

```python
# Retrieve EGFR sequence from NCBI using Entrez
# Run a BLAST search
# Save the XML file
# Parse the XML
# Print:Query name, Query length, Number of alignments
# Analyse the top hit: Title, Score, E-value, Alignment length, Identity %
# Loop through the first five hits
# Create the CSV file
```

```python
# 1. Retrieving EGFR sequence from NCBI Entrez

from Bio import Entrez, SeqIO

Entrez.email = "zaira.hasan656@gmail.com"

handle = Entrez.efetch(db = "nucleotide", id = "NM_005228.5", rettype = "fasta", retmode = "text")

record = SeqIO.read (handle, "fasta")
seq = record.seq

print("ID:", record.id)
print("Length of sequence:", len(seq))
print("First 100 bases:", seq[:100])
```

```python
# 2. Running BLAST search

from Bio.Blast import NCBIWWW
result_handle = NCBIWWW.qblast("blastn", "nt", seq)
```

```python
# 3. Saving the XML file

with open("EGFR_blast.xml", "w") as file:
  file.write(result_handle.read())

print("BLAST result saved.")
```

```python
# 4. Parsing XML and printing:Query name, Query length, Number of alignments

from Bio.Blast import NCBIXML

with open("EGFR_blast.xml") as result_handle:
  blast_result = NCBIXML.read(result_handle)

print(blast_result.application)
print(blast_result.query)
print(blast_result.query_length)
print(blast_result.alignments)
print(len(blast_result.alignments))
```

```python
# 5. Analyse the top hit: Title, Score, E-value, Alignment length, Identity %

top_hit = blast_result.alignments[0]
hsp = top_hit.hsps[0]

print("Name of the best sequence:", top_hit.title)
print("Length:", top_hit.length)
print("Score:", hsp.score)
print("E-value:", hsp.expect)
print("Alignment length:", hsp.align_length)
print("Identity:", round(hsp.identities/ hsp.align_length * 100,2),"%")
```

```python
# 6. Loop through the first five hits

for alignment in blast_result.alignments[:5]:
  hsp = alignment.hsps[0]

  print("Title:", alignment.title)
  print("Subject length:", alignment.length)
  print("Alignment length:", hsp.align_length)
  print("Score:", hsp.score)
  print("E-value:", hsp.expect)
  print("Identity:",round(hsp.identities/hsp.align_length * 100, 2), "%")
  print("-" * 60)
```

```python
# 7. Creating CSV file

import csv

with open("blast_result.csv", "w", newline="") as file:

    writer = csv.writer(file)

    writer.writerow([
        "Hit",
        "Title",
        "Identity %",
        "E-value",
        "Score",
        "Alignment Length"
    ])

    for i, alignment in enumerate(blast_result.alignments[:10], start=1):

        hsp = alignment.hsps[0]
        identity = (hsp.identities / hsp.align_length) * 100

        writer.writerow([
            i,
            alignment.title,
            round(identity, 2),
            hsp.expect,
            hsp.score,
            hsp.align_length
        ])

print("CSV report saved!")
```

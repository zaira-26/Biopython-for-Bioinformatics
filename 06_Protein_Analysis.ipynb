```python
from Bio.SeqUtils.ProtParam import ProteinAnalysis
```

```python
# Molecular Weight

# Molecular weight represents the combined mass of all amino acids within a protein sequence.
# It is commonly used in protein purification, electrophoresis and mass spectrometry.
```

```python
protein = "MVHLTPEEKSAVTALWGKV"
analysis = ProteinAnalysis(protein)
print("Protein:", protein)
print("Length of protein:", len(protein))
print("Molecular weight:", analysis.molecular_weight())
print("Amino acid count:", analysis.count_amino_acids())
print("Amino acid percentage:", analysis.amino_acids_percent)
```

```python
# Isoelectric point:
# The isoelectric point (pI) is the pH at which a protein carries no net electrical charge.
# It is useful in protein purification techniques such as isoelectric focusing.

# ph = pI : neutral
# pH > pI : net negative charge
# pH < pI : net positive charge
```

```python
protein = "MKWVTFISLLFLFSSAYSRGVFRRDTHKSEIAHRFKDLGE"

analysis = ProteinAnalysis(protein)
print("Protein:", protein)
print("Length:", len(protein))
print("Molecular weight of protein:", analysis.molecular_weight())
print("Amino acid count:", analysis.count_amino_acids())
print("Amino acid percentage:", analysis.amino_acids_percent)
print("Isoelectric point:", analysis.isoelectric_point())

highest_count = 0
most_abundant = ""
counts = analysis.count_amino_acids()
for amino_acids, count in counts.items():
  if count > highest_count:
    highest_count = count
    most_abundant = amino_acids

print("Most abundant protein:", most_abundant)
print("Frequency:", highest_count)
```

```python
protein = "MSDVEKGKKIFIMKCSQCHTVEKGGKHKTGPNEKSAA"
analysis = ProteinAnalysis(protein)
print("Protein:", protein)
print("Protein length:", len(protein))
print("Molecular weight of the protein:", analysis.molecular_weight())
print("Isoelectric point of the protein:", analysis.isoelectric_point())
print("Amino acid count:", analysis.count_amino_acids())
print("Amino acid percentage:", analysis.amino_acids_percent)

least_count = 100
least_common = ""
counts = analysis.count_amino_acids()
for amino_acids, count in counts.items():
  if count > 0 and count < least_count:
    least_count = count
    least_common = amino_acids

print("Least common amino acid:", least_common)
print("Frequency:", least_count)

single_amino_acids = []
counts = analysis.count_amino_acids()
for amino_acid, count in counts.items():
  if count == 1:
    single_amino_acids.append(amino_acid)

print("Single amino acid:", single_amino_acids)

```

```python
# Instability Index:
# The instability index estimates whether a protein is likely to remain stable under laboratory conditions. Proteins with an instability index below 40 are generally considered stable.

# If Instability index < 40 : Predicted stable
# If Instability index >= 40 : Predicted unstable
```

```python
# Aromaticity:
# Aromaticity gives the fraction of aromatic amino acids present in the protein.

# The 3 aromatic amino acids are phenylalanine, tyrosine and tryptophan.
```

```python
protein = "MKWVTFISLLFLFSSAYSRGVFRRDTHKSEIAHRFKDLGE"
analysis = ProteinAnalysis(protein)
print("Length of protein:", len(protein))
print("Molecular weight of protein:", analysis.molecular_weight())
print("Aromaticity:", analysis.aromaticity())
print("Isoelectric point of protein:", analysis.isoelectric_point())
print("Instability index of protein:", analysis.instability_index())

index = analysis.instability_index()
if index < 40:
  print("Predicted: Stable protein")
else:
  print("Predicted: Unstable protein")
```

```python
# GRAVY - Grand average of hydropathy
# GRAVY estimates the overall hydrophobicity or hydrophilicity of a protein.

# if GRAVY > 0 = hydrophobic
# if GRAVY < 0 = hydrophilic
```

```python
protein = "MSDVEKGKKIFIMKCSQCHTVEKGGKHKTGPNEKSAA"
analysis = ProteinAnalysis(protein)
print("Length of protein:", len(protein))
print("Molecular weight:", analysis.molecular_weight())
print("Protein structure probability:", analysis.secondary_structure_fraction())
print("Isoelectric point:", analysis.isoelectric_point())
print("Instability index:", analysis.instability_index())
print("Aromaticity:", analysis.aromaticity())
print("GRAVY", analysis.gravy())

if analysis.gravy() > 0:
  print("It is a hydrophobic protein.")

else:
  print("It is a hydrophilic protein.")
```

```python
# Secondary Structure Fraction:

# Biopython estimates the fraction of amino acids likely to form alpha helices, beta sheets and turns within a protein.
```

```python
protein = "MSPQTETKASVGFKAGVKEYKLTYYTPEYETKDTDILAAFRVTPQPGVPPEEAGAAVAAESSTGTWTTVWTDGLTSLDRYKGRCYHIEPVPG"
analysis = ProteinAnalysis(protein)
print("Length of protein:", len(protein))
print("Molecular weight of protein:", analysis.molecular_weight())
print("Amino acid count:", analysis.count_amino_acids())

highest_count = 0
most_abundant = ""
counts = analysis.count_amino_acids()
for amino_acids, count in counts.items():
  if count > highest_count:
    highest_count = count
    most_abundant = amino_acids

print("Most abundant amino acid is:", most_abundant)
print("Frequency:", highest_count)

least_count = 100
least_common = ""
counts = analysis.count_amino_acids()
for amino_acids, count in counts.items():
  if count > 0 and count < least_count:
    least_count = count
    least_common = amino_acids

print("Least abundant amino acid present in protein is:", least_common)
print("Frequency:", least_count)

print("Amino acid percentage:", analysis.amino_acids_percent)
print("Isoelectric point of protein:", analysis.isoelectric_point())
print("Instability index of protein:", analysis.instability_index())

if analysis.instability_index() < 40:
  print("Predicted: Stable protein")
else:
  print("Predicted: Unstable protein")

print("Aromaticity:", analysis.aromaticity())
print("GRAVY:", analysis.gravy())

if analysis.gravy() > 0:
  print("It is a hydrophobic protein")
else:
  print("It is a hydrophilic protein")

helix, turn, sheet = analysis.secondary_structure_fraction()
print("Helix:", helix)
print("Turns:", turn)
print("Beta sheets:", sheet)

if helix > turn and helix > sheet:
  print("Dominant predicted structure: Alpha helix")
elif turn > helix and turn > sheet:
  print("Dominant predicted structure: Turn")
else:
  print("Dominant predicted structure: Beta sheet")


```
